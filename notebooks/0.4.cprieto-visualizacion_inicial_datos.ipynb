```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# !pip install matplotlib
# Scatter plot (diagramas de dispersion)
from matplotlib.pyplot import title


x = np.arange(1,20)
y = x**2
y1 = x**2 + 5*x

plt.scatter(x,y,color='red',marker='s',label='x^2')
plt.scatter(x,y1,color='blue',marker='o',label='x^2 + 5')
plt.legend(title='Funciones')
plt.title('Tipos de Funciones', fontsize=20)
plt.xlabel('x')   # Nombre de la variable x
plt.ylabel('f(x)') # Nombre de la variable y
plt.grid()  # cuadricula

plt.text(5,300, 'este texto', color='brown')

# plt.xlim(0,5) para cambiar los datos de la variable x

# Agregar las coordenadas de los marcadores en texto
for val_x in x:
    plt.text(val_x,val_x**2, f'({val_x},{val_x**2})')
```

```python
# Histogramas
media = 20
std = 3
numeros = np.random.normal(media, std, 1000)

plt.hist(numeros, bins=20, cumulative=True, density=False);
plt.title('Distribucion Normal')
plt.xlabel('Numeros')
plt.ylabel('Frecuencia')
```
